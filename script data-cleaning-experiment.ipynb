###  I'll use Deterministic Regression impute to try to fill the missing values. If you want purity of data, you can skip and download the data we have cleaned so far . 

```python
# This is random impuration , it does not concern with relationship with other columns when filling missing values. 

missing_columns = ["beds","bath","size_sq.ft"]


def random_imputation(df, feature):

    number_missing = df[feature].isnull().sum()
    observed_values = df.loc[df[feature].notnull(), feature]
    df.loc[df[feature].isnull(), feature + '_imp'] = np.random.choice(observed_values, number_missing, replace = True)
    
    return df



for feature in missing_columns:
    df[feature + '_imp'] = df[feature]
    df = random_imputation(df, feature)


df.info()
```
